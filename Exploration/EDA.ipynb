### Customers vis-à-vis Subscribers: An Exploration

```python
gb_start = trips.groupby(by = ["Start Station", "Subscription Type"])
gb_end = trips.groupby(by = ["End Station", "Subscription Type"])
start_counts = pd.DataFrame(gb_start.count().ix[:, 0].copy())
end_counts = pd.DataFrame(gb_end.count().ix[:, 0].copy())
```

```python
start_counts.reset_index(inplace= True)
start_counts = start_counts.rename(columns = {"Start Station" : "Station" , "Trip ID" : "Count"})
```

```python
def normalize_col(df, to_normalize, to_subset, subset):
    ind = df[to_subset] == subset
    denom = df.ix[ind, [to_normalize]].sum()
    df.ix[ind, [to_normalize]] /= float(denom)
    return

normalize_col(start_counts, "Count", "Subscription Type", "Customer")
normalize_col(start_counts, "Count", "Subscription Type", "Subscriber")
```

```python
start_counts.head(4)
```

```python
subsc_ind = start_counts["Subscription Type"] == "Subscriber"
cust_ind = start_counts["Subscription Type"] == "Customer"

top_ten_subsc = start_counts.Station.loc[start_counts.loc[subsc_ind].Count.nlargest(10).index]
top_ten_cust = start_counts.Station.loc[start_counts.loc[cust_ind].Count.nlargest(10).index]

top_station_names = np.union1d(top_ten_subsc, top_ten_cust)
top_station_ind = np.union1d(top_ten_subsc.index, top_ten_cust.index)

top_counts = start_counts[start_counts["Station"].isin(top_station_names)]
top_counts.reset_index(inplace = True, drop = True)
```

```python
n_groups = top_station_names.shape[0] # 14
index = np.arange(n_groups)[::-1]

bar_width = 0.35
sns.set_style("dark")

plt.figure(figsize = (16, 12))

first = plt.barh(index, top_counts.Count[cust_ind], 
                 height = bar_width, 
                 label = "Customers", color = "indianred")

second = plt.barh(index + bar_width, top_counts.Count[subsc_ind], 
                   height = bar_width, label = "Subscribers")

plt.yticks(index + bar_width, top_station_names, fontsize = 14)
plt.title("Distribution of Start Desintations", fontsize = 14)
plt.legend(fontsize = 14);
```

**Takeaways**

* Customers are much more likely to visit touristey places like the Ferry Building and Embarcadero
* Subscribers are more likely to be at commuter hubs, indicating that they probably do not live in San Francisco

```python
subscribers = trips["Subscription Type"] == "Subscriber"
customers = trips["Subscription Type"] == "Customer"

starts_subsc = pd.DataFrame(data = trips.ix[subscribers, ["Trip ID", "Start Date"]])
starts_subsc.set_index("Start Date", inplace = True)
starts_subsc["Trip ID"] = 1
starts_subsc = starts_subsc["Trip ID"]

starts_cust = pd.DataFrame(data = trips.ix[customers, ["Trip ID", "Start Date"]])
starts_cust.set_index("Start Date", inplace = True)
starts_cust["Trip ID"] = 1
starts_cust = starts_cust["Trip ID"]


pd.stats.moments.rolling_mean(starts_cust.resample("1D", "sum"), window = 5, center = True).plot(figsize = (16, 8), 
                                                                               label = "Customers");
pd.stats.moments.rolling_mean(starts_subsc.resample("1D", "sum"), window = 5, center = True).plot(secondary_y = False,
                                                                                label = "Subscribers");
plt.legend(loc = 2);
plt.title("EWMA of Departures", fontsize = 14);
```
